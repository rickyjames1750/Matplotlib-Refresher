```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='green',linewidth=5,linestyle='dashdot')
```

```python
plt.plot(x,y, 'rD')
```

```python
plt.plot(x,y,color='red',marker='+',linestyle='',markersize='20')
```

```python
plt.plot(x,y,color='green',alpha=1)
```

```python
days = [1,2,3,4,5,6,7]
max_t = [90,98,92,97,99,100,91]
min_t = [43,42,40,30,36,38,20]
avg_t = [50,52,53,54,56,57,58]
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(days,max_t, label="Max")
plt.plot(days,min_t, label="Min")
plt.plot(days,avg_t, label="Avg")

plt.legend(loc="best", shadow="True", fontsize='small')

plt.grid()
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.xlabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")

plt.legend()
```

```python
plt.yticks(xpos,company)
plt.title("US Tech Stocks")
plt.barh(xpos-0.2,revenue,label="Revenue")
plt.barh(xpos+0.2,profit,label="Profit")

plt.legend()
```

```python
blood_sugar_humans = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_aliens = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood Sugar Analysis')
plt.hist([blood_sugar_humans, blood_sugar_aliens], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['humans','aliens'],orientation='horizontal')
plt.legend()
```
